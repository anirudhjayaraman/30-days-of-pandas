```python
import pandas as pd

def bonus(employee_id: int, name: str, salary: int) -> int:
    if((employee_id % 2 != 0) & (name[0] != 'M')):
        return(salary)
    else:
        return(0)

def calculate_special_bonus(employees: pd.DataFrame) -> pd.DataFrame:
    employees['bonus'] = employees.apply(func = lambda row: bonus(row['employee_id'], row['name'], row['salary']), axis = 1)
    return employees[['employee_id','bonus']].sort_values(by = 'employee_id', ascending = True)
    
```

```python
df1 = pd.DataFrame({'employee_id':[2,3,7,9,8],
                   'name':['Meir','Michael','Addilyn','Juan','Kannon'],
                   'salary':[3000,3800,7400,6100,7700]})

df1
```

```python
calculate_special_bonus(df1)
```
