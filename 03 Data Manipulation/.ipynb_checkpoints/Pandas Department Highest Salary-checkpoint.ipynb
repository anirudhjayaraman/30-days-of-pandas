```python
import pandas as pd

def department_highest_salary(employee: pd.DataFrame, department: pd.DataFrame) -> pd.DataFrame:
    joined = pd.merge(left = employee.rename(columns = {'name':'Employee','salary':'Salary'}), 
                   right = department.rename(columns = {'id':'departmentId','name':'Department'}),how='left', on='departmentId')
    return joined.loc[joined.Salary == joined.groupby('Department')['Salary'].transform('max'),['Department','Employee','Salary']]
```

```python
data = [[1, 'Joe', 70000, 1], [2, 'Jim', 90000, 1], [3, 'Henry', 80000, 2], [4, 'Sam', 60000, 2], [5, 'Max', 90000, 1]]
df_employee = pd.DataFrame(data, columns=['id', 'name', 'salary', 'departmentId']).astype({'id':'Int64', 'name':'object', 'salary':'Int64', 'departmentId':'Int64'})
data = [[1, 'IT'], [2, 'Sales']]
df_department = pd.DataFrame(data, columns=['id', 'name']).astype({'id':'Int64', 'name':'object'})
```
