```python
import pandas as pd

def order_scores(scores: pd.DataFrame) -> pd.DataFrame:
    return pd.merge(left = scores.sort_values(by = ['score'], ascending = False).loc[:,['score']],
        right = pd.DataFrame(data = {'score':sorted(list(set(scores.score))),
                     'rank':[i for i in range((len(set(scores.score))),0,-1)]}),
        how = 'left',
        on = 'score')
```

```python
data = [[1, 3.5], [2, 3.65], [3, 4.0], [4, 3.85], [5, 4.0], [6, 3.65]]
scores = pd.DataFrame(data, columns=['id', 'score']).astype({'id':'Int64', 'score':'Float64'})
```
