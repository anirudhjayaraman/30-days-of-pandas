```python
import pandas as pd

def ads_performance(ads: pd.DataFrame) -> pd.DataFrame:
    df1 = ads.groupby(['ad_id','action']).agg('count').reset_index().rename(columns = {'user_id':'count'})
    df2 = ads.loc[ads.action != "Ignored", ['ad_id','action']].groupby(by = 'ad_id').agg('count').reset_index().rename(columns = {'action':'total'})
    res = pd.merge(left=df1.loc[df1.action == 'Clicked',['ad_id','count']], right = df2, on = 'ad_id')
    res['ctr'] = round(100 * res['count'] / res['total'],2)

    try:
        return pd.concat([res[['ad_id','ctr']], pd.DataFrame(data = [[el,0] for el in list(ads.ad_id.unique()) if el not in list(res.ad_id)], columns = ['ad_id','ctr'])], axis = 0).sort_values(['ctr','ad_id'], ascending = [False, True])
    except:
        return res[['ad_id','ctr']].sort_values(['ctr','ad_id'], ascending = [False, True])
```

```python
data = [[1, 1, 'Clicked'], [2, 2, 'Clicked'], [3, 3, 'Viewed'], [5, 5, 'Ignored'], [1, 7, 'Ignored'], [2, 7, 'Viewed'], [3, 5, 'Clicked'], [1, 4, 'Viewed'], [2, 11, 'Viewed'], [1, 2, 'Clicked']]
ads = pd.DataFrame(data, columns=['ad_id', 'user_id', 'action']).astype({'ad_id':'Int64', 'user_id':'Int64', 'action':'object'})
```

```python
ads_performance(ads)
```
