```python
import pandas as pd

def categorize_products(activities: pd.DataFrame) -> pd.DataFrame:
    activities.drop_duplicates(inplace=True)
    df1 = activities.groupby(['sell_date']).agg('count').reset_index().rename(columns = {'product':'num_sold'})
    df1['products'] = pd.NA
    for i in range(df1.shape[0]):
        df1.iloc[i,2] = ','.join(sorted(activities.loc[activities.sell_date == df1.sell_date[i],'product'].to_list()))
    return df1
```

```python
data = [['2020-05-30', 'Headphone'], ['2020-06-01', 'Pencil'], ['2020-06-02', 'Mask'], ['2020-05-30', 'Basketball'], ['2020-06-01', 'Bible'], ['2020-06-02', 'Mask'], ['2020-05-30', 'T-Shirt']]
activities = pd.DataFrame(data, columns=['sell_date', 'product']).astype({'sell_date':'datetime64', 'product':'object'})
activities
```

```python
categorize_products(activities)
```
