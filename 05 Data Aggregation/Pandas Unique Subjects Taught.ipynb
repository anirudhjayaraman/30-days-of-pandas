```python
import pandas as pd

def count_unique_subjects(teacher: pd.DataFrame) -> pd.DataFrame:
    return teacher[['teacher_id','subject_id']].drop_duplicates().groupby(['teacher_id']).agg('count').reset_index().rename(columns = {'subject_id':'cnt'})
```

```python
data = [[1, 2, 3], [1, 2, 4], [1, 3, 3], [2, 1, 1], [2, 2, 1], [2, 3, 1], [2, 4, 1]]
teacher = pd.DataFrame(data, columns=['teacher_id', 'subject_id', 'dept_id']).astype({'teacher_id':'Int64', 'subject_id':'Int64', 'dept_id':'Int64'})
teacher
```

```python
teacher[['teacher_id','subject_id']].drop_duplicates().groupby(['teacher_id']).agg('count').reset_index().rename(columns = {'subject_id':'cnt'})
```

```python
count_unique_subjects(teacher)
```
